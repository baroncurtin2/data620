```python
# imports and config
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create dictionary of nodes and edges
setup = {
    'Jane': ['Ike'],
    'Ike': ['Heather', 'Jane'],
    'Heather': ['Fernando', 'Garth', 'Ike'],
    'Fernando': ['Garth', 'Diane', 'Andre', 'Carol'],
    'Garth': ['Heather', 'Fernando', 'Diane', 'Ed', 'Beverly'],
    'Carol': ['Fernando', 'Diane', 'Andre'],
    'Diane': ['Carol', 'Fernando', 'Garth', 'Ed', 'Beverly', 'Andre'],
    'Ed': ['Garth', 'Diane', 'Beverly'],
    'Andre': ['Carol', 'Diane', 'Beverly', 'Fernando'],
    'Beverly': ['Andre', 'Diane', 'Garth', 'Ed']
}

setup = {node: [(node, e) for e in edges] for node, edges in setup.items()}
```

```python
# create networkx graph
G = nx.Graph()
G.add_nodes_from(setup.keys())
G.add_edges_from([y for x in setup.values()
                  for y in x])

nx.draw(G, with_labels=True)
```
