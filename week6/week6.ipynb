```python
import matplotlib.pyplot as plt
import networkx as nx
import networkx.algorithms.bipartite as bipartite
```

```python
G = nx.davis_southern_women_graph()
women = G.graph['top']
clubs = G.graph['bottom']

print("Biadjacency matrix")
print(bipartite.biadjacency_matrix(G, women, clubs))
```

```python
# project bipartite graph onto women nodes
W = bipartite.projected_graph(G, women)
print('')
print("#Friends, Member")
for w in women:
    print('%d %s' % (W.degree(w), w))
```

```python
# project bipartite graph onto women nodes keeping number of co-occurence
# the degree computed is weighted and counts the total number of shared contacts
W = bipartite.weighted_projected_graph(G, women)
print('')
print("#Friend meetings, Member")
for w in women:
    print('%d %s' % (W.degree(w, weight='weight'), w))
```

```python
labels = {w: ''.join([x[0] for x in w.split(' ')]) for w in women}

G = nx.relabel_nodes(G, labels)
nx.draw(G, with_labels=True, node_color=['blue' if node in clubs else 'green' 
                                         for node in G.nodes()])
plt.show()
    
```

From there graph, we can infer that the women that go to the same events are probably friends or connected in some way.
Some other inferences we can make are the women that go to move events are likely to know more women than the ones that do not go to as many events.
All of the women are somehow connected to each other. There are no disconnected edges/nodes which means that if a connection needs to be made, there are people that can direct the connection.
